```python
import numpy as np
import pandas as pd
import cv2
import os
```

```python
# CODE FOR CHECKING NAN OCCURRENCES # 

#REPLACE
###
files_end_with = '_normalized.csv'
one_file = False #True: Analize one file (replace excel_path); False: Analize all the files in a folder (replace folder_path): 
excel_path = r"C:\Users\catar\OneDrive\Ambiente de Trabalho\champalimaud\new"
folder_path = r'C:\Users\catar\OneDrive\Ambiente de Trabalho\champalimaud\finalfiles'
###

dataframes = []
if one_file == True: 
    dataframes.append(pd.read_csv(excel_path))
else: 
    file_list = os.listdir(folder_path)
    csv_files = [file for file in file_list if (file.endswith(files_end_with))] 
    
    for csv_file in csv_files:
        file_path = os.path.join(folder_path, csv_file)
        df = pd.read_csv(file_path)
        dataframes.append(df)


nan_head = []
nan_infer = []
nan_abdomen = []
nan_leftwing = []
nan_rightwing = []
nan_thorax = []
nan_centroid =[]

for data in dataframes:
    df = data[['Centroid.X', 'Centroid.Y', 'InferPosition.X', 'InferPosition.Y', 'Abdomen.Position.X','Abdomen.Position.Y', 'LeftWing.Position.X', 'LeftWing.Position.Y', 'RightWing.Position.X', 'RightWing.Position.Y', 'Head.Position.X', 'Head.Position.Y', 'Thorax.Position.X', 'Thorax.Position.Y']]
    nan_head.append(np.isnan(df['Head.Position.X']).sum())
    nan_infer.append(np.isnan(df['InferPosition.X']).sum())
    nan_abdomen.append(np.isnan(df['Abdomen.Position.X']).sum())
    nan_leftwing.append(np.isnan(df['LeftWing.Position.X']).sum())
    nan_rightwing.append(np.isnan(df['RightWing.Position.X']).sum())
    nan_thorax.append(np.isnan(df['Thorax.Position.X']).sum())
    nan_centroid.append(np.isnan(df['Centroid.X']).sum())

print("Head", nan_head)
print("Abdomen", nan_abdomen)
print("Infer", nan_infer)
print("Left Wing", nan_leftwing)
print("Right Wing", nan_rightwing)
print("Thorax", nan_thorax)
print("Centroid", nan_centroid)

# Missed Head too many times (500), go to trash: 
n = 0
while n < len(dataframes):
    if nan_head[n] > 500: 
        if one_file == True:
            output_trash = os.path.dirname(excel_path)
            output_trash_path = output_trash + '/Trash/'
            os.makedirs(output_trash_path, exist_ok=True) #create trash folder if it doesn't exist
            dataframes[n].to_csv(f'{output_trash_path}/{os.path.basename(excel_path)}', index=False)
            #os.remove(excel_path)                                                                     #IF YOU WANT TO DELETE THE FILE FROM THE ORIGINAL FOLDER, UNCOMMENT
        else: 
            print("Number of NaNs in the faulty files:", nan_head[n])
            output_trash_path = folder_path + '/Trash/'
            os.makedirs(output_trash_path, exist_ok=True) #create trash folder if it doesn't exist  
            dataframes[n].to_csv(f'{output_trash_path}/{csv_files[n]}', index=False)
            os.remove(folder_path + '\\' + csv_files[n])                                               #IF YOU WANT TO DELETE THE FILE FROM THE ORIGINAL FOLDER, UNCOMMENT

    n = n + 1 
```

```python
# CODE FOR GETTING VIDEO WITH COORDINATES MARKED #

#REPLACE:
###
excel_path = r"C:\Users\catar\OneDrive\Ambiente de Trabalho\champalimaud\new\VLF-39347_DH44_RNAi_TRIP-5m_base_5stim_20loom-3_4days-FH1_1-2023-12-21T21_01_53_fly1_infered_sleap_tracked_normalized.csv"
video_path = r"C:\Users\catar\OneDrive\Ambiente de Trabalho\champalimaud\new\VLF-39347_DH44_RNAi_TRIP-5m_base_5stim_20loom-3_4days-FH1_1-2023-12-21T21_01_53_fly1.avi"
output_path = r"C:\Users\catar\OneDrive\Ambiente de Trabalho\champalimaud\videotestevioletta2.avi"
###


cap = cv2.VideoCapture(video_path)

file = pd.read_csv(excel_path)
df = file[['Centroid.X', 'Centroid.Y',  'Abdomen.Position.X','Abdomen.Position.Y', 'LeftWing.Position.X', 'LeftWing.Position.Y', 'RightWing.Position.X', 'RightWing.Position.Y', 'Head.Position.X', 'Head.Position.Y', 'Thorax.Position.X', 'Thorax.Position.Y']]
coordinates = np.nan_to_num(df.values, nan=0)

#OUTPUT
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*'XVID')  

out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))


while True:
    ret, frame = cap.read() #get next frame
    
    #EXIT at the end of the video
    if not ret:
        break

    frame_number = int(cap.get(cv2.CAP_PROP_POS_FRAMES)) #Current frame
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT)) #Total frames
    
    frame_coordinates = None  

    if frame_number < total_frames:
        frame_coordinates = coordinates[frame_number - 1] #Start at index 0
    
      
    colors = [(0, 255, 0), (255, 0, 0), (0, 0, 255), (255, 255, 0), (255, 255, 255), (255,0,255)]
    
    if frame_coordinates is not None:
        
        #Putting dots with different colors in the video
        f = 0
        for i in range(0, len(frame_coordinates), 2):
            x, y = map(int, frame_coordinates[i:i+2])
            color = colors[f]
            f = f + 1
            cv2.circle(frame, (x, y), 2, color, -1)  

    #Output video
    out.write(frame)

```
